```julia
function draw(M)
    x=M[:,1]
    y=M[:,2]
    z=x+im*y
    e=z.^2
    #print(norm(e[1]+e[2]+e[3]),"\n")
    #print(norm(e[1])+norm(e[2])+norm(e[3]))
    E=[real(e)' ; imag(e)']
    points=[zeros(2,1) E[:,1] E[:,1] E[:,1]+E[:,2] E[:,1]+E[:,2] E[:,1]+E[:,2]+E[:,3] E[:,1]+E[:,2]+E[:,3]]
    return points
end
```

```julia
points=draw(E1)
plot(points[1,:],points[2,:])
```
